(<a href='#Table of Contents'>Top</a>)<a id='boost'></a><h3>Install boost library</h3>
Sassena uses an older version of boost

```python
#!/bin/bash
module load openmpi-x86_64
PREFIX=/sw/sassena/1.4.1
cd $PREFIX
tar -xzf boost_1_47_0.tar.gz
cd $PREFIX/boost_1_47_0
```

There are required changes to a few of the source files in  boost 1.47 prior to compiling, <i>if you are compiling with gcc > 4.6</i>:

Edit file ./boost/config/stdlib/libstdcpp3.hpp and substitute line 36:  
<code>|| defined(_GLIBCXX__PTHREADS)</code>  
with these two lines  
<code>|| defined(_GLIBCXX__PTHREADS) \
|| defined(_GLIBCXX_HAS_GTHREADS)</code>

Substitute <code>TIME\_UTC</code> with <code>TIME\_UTC\_</code> in all files with the following command:  
<code>for file in `grep -R TIME_UTC * |cut -d ':' -f 1|sort -u|tr '\n' ' '`;do
   perl -pi -e 's/TIME\_UTC/TIME\_UTC\_/g' $file
done</code>

After these changes we can compile safely.

```python
module load openmpi-x86_64  # Load the environment for mpi compilation
PREFIX=/sw/sassena/1.4.1
TARGETDIR=/sw/sassena/1.4.1
mkdir -p $TARGETDIR/include/boost_1_47_0
mkdir -p $TARGETDIR/lib64/boost_1_47_0
cd $PREFIX/boost_1_47_0
./bootstrap.sh --includedir=$TARGETDIR/include/boost_1_47_0 --libdir=$TARGETDIR/lib64/boost_1_47_0
echo "using mpi ; " >> project-config.jam
./b2 install
```

The previous commands will compile boost and save artifacts to the include and lib64 directories

(<a href='#Table of Contents'>Top</a>)<a id='fftw'></a><h3>Install FFTW3 library</h3>  
Library for fast fourier transform

```python
module load openmpi-x86_64
PREFIX=/sw/sassena/1.4.1
cd $PREFIX/
tar -xzf fftw-3.3.tar.gz
cd $PREFIX/fftw-3.3
./configure --prefix=$PREFIX
make && make install
./configure --prefix=$PREFIX --enable-shared
make clean && make && make install
```

(<a href='#Table of Contents'>Top</a>)<a id='sassena'></a><h3>Install Sassena</h3>  
In addition to <code>boost</code> and <code>fftw</code>, sassena requires knowing the location of other libraries:  

* BLAS
* LAPACK
* HDF5

Here we provide in painful detail of how to export all neccessary environment variables to compile in cluster fermi.ornl.gov. Many of these variables are commented out, but you may need to uncomment some if your environment is unable to find them:

```python
module load openmpi-x86_64
PREFIX=/sw/fermi/sassena/1.4.1
cd $PREFIX
tar -zxf sassena-v1.4.1.tar.gz

#######################
#  Boost environment  #
#######################
BOOST_ROOT=$PREFIX/boost_1_47_0
Boost_INCLUDE_DIR=${BOOST_ROOT}
BOOST_LIBRARYDIR=$PREFIX/lib64/boost_1_47_0
Boost_LIBRARY_DIRS=$BOOST_LIBRARYDIR
Boost_FILESYSTEM_LIBRARY=$BOOST_LIBRARYDIR/libboost_filesystem.so
Boost_FILESYSTEM_LIBRARY_DEBUG=$BOOST_LIBRARYDIR/libboost_filesystem.so
DBoost_FILESYSTEM_LIBRARY_RELEASE=$BOOST_LIBRARYDIR/libboost_filesystem.so
Boost_SYSTEM_LIBRARY=$BOOST_LIBRARYDIR/libboost_system.so
Boost_SYSTEM_LIBRARY_DEBUG=$BOOST_LIBRARYDIR/libboost_system.so
Boost_SYSTEM_LIBRARY_RELEASE=$BOOST_LIBRARYDIR/libboost_system.so
Boost_REGEX_LIBRARY=$BOOST_LIBRARYDIR/libboost_regex.so
Boost_REGEX_LIBRARY_DEBUG=$BOOST_LIBRARYDIR/libboost_regex.so
Boost_REGEX_LIBRARY_RELEASE=$BOOST_LIBRARYDIR/libboost_regex.so
Boost_THREAD_LIBRARY=$BOOST_LIBRARYDIR/libboost_thread.so
Boost_THREAD_LIBRARY_DEBUG=$BOOST_LIBRARYDIR/libboost_thread.so
Boost_THREAD_LIBRARY_RELEASE=$BOOST_LIBRARYDIR/libboost_thread.so
Boost_PROGRAM_OPTIONS_LIBRARY=$BOOST_LIBRARYDIR/libboost_program_options.so
Boost_PROGRAM_OPTIONS_LIBRARY_DEBUG=$BOOST_LIBRARYDIR/libboost_program_options.so
Boost_PROGRAM_OPTIONS_LIBRARY_RELEASE=$BOOST_LIBRARYDIR/libboost_program_options.so
Boost_SERIALIZATION_LIBRARY=$BOOST_LIBRARYDIR/libboost_serialization.so
Boost_SERIALIZATION_LIBRARY_DEBUG=$BOOST_LIBRARYDIR/libboost_serialization.so
Boost_SERIALIZATION_LIBRARY_RELEASE=$BOOST_LIBRARYDIR/libboost_serialization.so

#####################
# BLAS environment  #
#####################
#BLASLIB=acml
#BLAS_ROOT=/cm/shared/apps/acml/5.3.1/gfortran64
#BLAS_DIR=$BLAS_ROOT/lib
#BLASDIR=$BLAS_ROOT/lib
#BLAS_LIBRARIES=$BLASDIR/libacml.so
#LD_LIBRARY_PATH=$BLASDIR:LD_LIBRARY_PATH
#INCLUDE=$BLAS_ROOT/include
#module load acml/gcc/64/5.3.1
#BLAS_LIBRARIES=/cm/shared/apps/acml/5.3.1/gfortran64/lib/libacml.so
BLAS_LIBRARIES=/usr/lib64/libblas.so

#######################
# LAPACK environment  #
#######################
#module load lapack
#LAPACK_LIBRARIES=/usr/lib64/liblapack.so
LAPACK_LIBRARIES=/usr/lib64/atlas/liblapack.so.3

#####################
# FFTW environment  #
#####################
export FFTW3_HOME=$PREFIX

#####################
# HDF5 environment  #
#####################
HDF5_INCLUDE_DIR=/usr/include
HDF5_LIBRARY_DIR=/usr/lib64
HDF5HL_LIBRARY_DIR=/usr/lib64
HDF5_LIBRARIES=${HDF5_LIBRARY_DIR}/libhdf5.so
HDF5_LIBRARY=${HDF5_LIBRARY_DIR}/libhdf5.so
HDF5HL_LIBRARY=${HDF5_LIBRARY_DIR}/libhdf5_hl.so
LD_LIBRARY_PATH=hdf5:$LD_LIBRARY_PATH

########################################
# Sassena environment and compilation  #
########################################

SASSENA_HOME=$PREFIX/sassena-v1.4.1
BUILDIR=$PREFIX/build
mkdir -p $BUILDIR && cd $BUILDIR

SASSENA_HOME=$PREFIX/sassena-v1.4.1
BUILDIR=$PREFIX/build
mkdir -p $BUILDIR && cd $BUILDIR

cmake -D CMAKE_EXE_LINKER_FLAGS="-Wl,-rpath,$LD_LIBRARY_PATH" -DCMAKE_INSTALL_PREFIX=$PREFIX \
$SASSENA_HOME -DBoost_INCLUDE_DIR=${Boost_INCLUDE_DIR} -DBOOST_LIBRARYDIR=${BOOST_LIBRARYDIR} \
-DBoost_LIBRARY_DIRS=${Boost_LIBRARY_DIRS} -DBoost_FILESYSTEM_LIBRARY=${Boost_FILESYSTEM_LIBRARY} \
-DBoost_FILESYSTEM_LIBRARY_DEBUG=${Boost_FILESYSTEM_LIBRARY_DEBUG=} \
-DBoost_FILESYSTEM_LIBRARY_RELEASE=${Boost_FILESYSTEM_LIBRARY_RELEASE} \
-DBoost_SYSTEM_LIBRARY=${Boost_SYSTEM_LIBRARY} \
-DBoost_SYSTEM_LIBRARY_DEBUG=${Boost_SYSTEM_LIBRARY_DEBUG} \
-DBoost_SYSTEM_LIBRARY_RELEASE=${Boost_SYSTEM_LIBRARY_RELEASE} \
-DBoost_REGEX_LIBRARY=${Boost_REGEX_LIBRARY} \
-DBoost_REGEX_LIBRARY_DEBUG=${Boost_REGEX_LIBRARY_DEBUG} \
-DBoost_REGEX_LIBRARY_RELEASE=${Boost_REGEX_LIBRARY_RELEASE} \
-DBoost_THREAD_LIBRARY=${Boost_THREAD_LIBRARY} \
-DBoost_THREAD_LIBRARY_DEBUG=${Boost_THREAD_LIBRARY_DEBUG} \
-DBoost_THREAD_LIBRARY_RELEASE=${Boost_THREAD_LIBRARY_RELEASE} \
-DBoost_PROGRAM_OPTIONS_LIBRARY=${Boost_PROGRAM_OPTIONS_LIBRARY} \
-DBoost_PROGRAM_OPTIONS_LIBRARY_DEBUG=${Boost_PROGRAM_OPTIONS_LIBRARY_DEBUG} \
-DBoost_PROGRAM_OPTIONS_LIBRARY_RELEASE=${Boost_PROGRAM_OPTIONS_LIBRARY_RELEASE} \
-DBoost_SERIALIZATION_LIBRARY=${Boost_SERIALIZATION_LIBRARY} \
-DBoost_SERIALIZATION_LIBRARY_DEBUG=${Boost_SERIALIZATION_LIBRARY_DEBUG} \
-DBoost_SERIALIZATION_LIBRARY_RELEASE=${Boost_SERIALIZATION_LIBRARY_RELEASE} \
-DHDF5_INCLUDE_DIR=${HDF5_INCLUDE_DIR} -DHDF5_LIBRARY_DIR=${HDF5_LIBRARY_DIR} \
-DHDF5HL_LIBRARY_DIR=${HDF5HL_LIBRARY_DIR} -DHDF5_LIBRARIES=${HDF5_LIBRARIES} \
-DHDF5_LIBRARY=${HDF5_LIBRARY} -DHDF5HL_LIBRARY=${HDF5HL_LIBRARY} -DCMAKE_BUILD_TYPE=Debug \
-DBLAS_LIBRARIES=${BLAS_LIBRARIES} -DLAPACK_LIBRARIES=${LAPACK_LIBRARIES} $SASSENA_HOME

make && make install
```

artifacts (shared libraries and executables) will be stored in <i>/sw/fermi/sassena/1.4.1/build</i>. At this point it is convenient to create a modules file to load the appropriate environment. In the case of fermi.ornl.gov, we save module file <code>1.4.1</code> under directory <code>/sw/fermi/modulefiles/sassena/</code>. The contents of the module file:  
<pre>#%Module
proc ModulesHelp { } {
  puts stderr "SASSENA"
  puts stderr ""
}

# One line description
module-whatis "Sassena 1.4.1, MPI"

module load openmpi-x86_64
set SASSENAHOME /sw/fermi/sassena/1.4.1
set SASSENADATABASE $SASSENAHOME/database

prepend-path LD_LIBRARY_PATH $SASSENAHOME/build
prepend-path PATH $SASSENAHOME/build:$SASSENAHOME/bin

setenv       SASSENA_DATABASE $SASSENADATABASE

puts stderr "Sassena 1.4.1 MPI. Executable is 'sassena'\nSASSENA_DATABASE=$SASSENADATABASE"
</pre>

And of course, we make sure that directory <code>/sw/fermi/modulefiles</code> is part of environment variable
<code>MODULEPATH</code>:  

<pre>fermi-login1:build> echo $MODULEPATH
/SNS/users/jbq/.local/modulefiles:/sw/fermi/modulefiles:/etc/modulefiles/:/usr/share/Modules/modulefiles:
</pre>

Also, we include file <code>.version</code> in <code>/sw/fermi/modulefiles/sassena/</code> with the following contents:

<pre>#%Module1.0
##
##  The desired version of EPW
##
set ModulesVersion "1.4.1"
</pre>

This file will set file <code>1.4.1</code> as the default module. With all this in place, we call the sassena module to load the right environment in our queue scripts. for instance:

<pre>
#!/bin/bash

#PBS -N sassenaInc
#PBS -j oe
#PBS -l walltime=12:00:00
#PBS -l nodes=4:ppn=16

#Request four nodes and using 16 cores per node 
source $MODULESHOME/init/bash
module load sassena

cd $PBS_O_WORKDIR
mpirun -np 64 sassena --config=sassenaInc.xml &> sassenaInc.log
</pre>
